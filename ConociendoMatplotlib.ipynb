```python
# Importamos numpy
import numpy as np

# Generamos valores aleatorios
np.random.seed(1000)

# Defnimos numeros aleatorios para los valores de y
y = np.random.standard_normal(20)

# Corregimos los numeros enteros de x
x = np.arange(len(y))

# Graficamos los valores de y x
plt.plot(x, y);
```
